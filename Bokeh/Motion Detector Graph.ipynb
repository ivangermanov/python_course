```python
from bokeh.plotting import figure, show, output_file
import pandas

output_file("rectangles.html")

df=pandas.read_csv("Times.csv")
i = 1

f=figure(width=400, height=400)

for datum in df["Start"]:
    f.quad(left=df["Start"], right=df["End"], bottom=i, top=i+1)
    i=i+1

show(f)
```
